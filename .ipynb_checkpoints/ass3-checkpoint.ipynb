```python
# for age-income dataset
data = {
    'Age' : [22,25,47,52,46,56,55,60,23,34,44,53],
     'Income': [25000, 48000, 52000, 58000, 60000, 62000, 61000, 70000, 26000, 40000, 52000, 58000],
    'AgeGroup': ['Young', 'Young', 'Middle-aged', 'Middle-aged', 'Middle-aged', 'Senior', 'Senior', 'Senior', 'Young', 'Young', 'Middle-aged', 'Senior']
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# Step 1a : grouped summary statistics

grouped_summary = df.groupby('AgeGroup')['Income'].agg(['mean','median','min','max','std'])
print('\n---- Grouped Summary Statstics (Income grouped by AgeGroup) ----\n')
print(grouped_summary)
```

```python
# Step 1b: list of Income values for each agegroup

grouped_list = df.groupby('AgeGroup')['Income'].apply(list).to_dict()
```

```python
print("\n--- List of Incomes per AgeGroup ---\n")
for group, values in grouped_list.items():
    print(f"{group} : {values}\n")
```
